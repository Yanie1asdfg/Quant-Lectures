# Fama-French五因子模型

早在1993年，Fama和French两个人就已经发表了他们的三因子模型，认为股票的超额收益可以由市场风险、市值风险、账面市值比风险来共同解释。后来，这两个人发现了除了上述风险，还有盈利水平风险、投资水平风险也能带来个股的超额收益，并在2013年发表了五因子模型。

在过去20年里面，很多学者对三因子模型进行实证分析，发现有些股票的αi显著不为0，这说明三因子模型中的三个风险（因素）并不能解释所有超额收益。
