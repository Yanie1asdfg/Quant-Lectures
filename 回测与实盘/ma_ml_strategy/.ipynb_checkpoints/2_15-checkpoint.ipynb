```python
# 查看持仓量的图 grag_volume,grag_open_interest
data['open_interest'].plot()
data[data['open_interest']>grag_open_interest]['open_interest'].plot(figsize=(12,8))
```

```python
# 查看成交量的图
data['volume'].plot()
data[data['volume']<grag_volume]['volume'].plot(figsize=(12,8))
```

```python
data1 = data.copy()
```

```python
for i in data1.index:
    if data1.loc[i,'open_interest']>grag_open_interest:
        start1 = i
        break
for i in data1.index:
    if data1.loc[i,'volume']>grag_volume:
        start2 = i
        break
start1,start2  # 起始的截断选择open_interest更好
```

```python
# 查看尾部持仓量的图 grag_volume,grag_open_interest
data['open_interest'].plot()
data[data['open_interest']>grag_open_interest2]['open_interest'].plot(figsize=(12,8))
```

```python
for i in data1.index[::-1]:
    if data1.loc[i,'volume']<grag_volume and data1.loc[i-1,'volume']>grag_volume:
        end1 = i
        break
for i in data1.index[::-1]:
    if data1.loc[i,'open_interest']<grag_open_interest and data1.loc[i-1,'open_interest']>grag_open_interest:
        end2 = i
        break
end1,end2 # open_interest更好
```

```python
data1.iloc[max(start1,start2):min(end1,end2),:]
```

* 所以选择 50%volume，40%open_interest的最大值作为起始较为合适
* 选择 50%volume，50%open_interest的最小值作为终止较为合适
* 最新数据尾部不用处理

### 处理所有数据

```python
import os
path = r'E:\repo\ml_bt'
all_path = [os.path.join(path,x) for x in os.listdir(path)]
all_path
```

```python
import matplotlib.pyplot as plt
for i in all_path:
    data = pd.read_csv(i,index_col=0)
    grag_volume= data.describe([0.1,0.25,0.3,0.4,0.5]).T.loc['volume','50%']
    grag_open_interest = data.describe([0.1,0.25,0.3,0.4,0.5]).T.loc['open_interest','40%']
    grag_open_interest2 = data.describe([0.1,0.25,0.3,0.4,0.5]).T.loc['open_interest','50%']
    print(grag_volume,grag_open_interest,grag_open_interest2)
    # 查看持仓量的图 grag_volume,grag_open_interest
    fig = plt.figure(figsize=(12,8))
    plt.plot(data.index,data['open_interest'])
    dt = data[data['open_interest']>grag_open_interest]
    plt.plot(dt.index,dt['open_interest'])
```

* 经观察发现，上述对于活跃成交量的处理可行

```python
for e in all_path:
    data = pd.read_csv(e,index_col=0)
    grag_volume= data.describe([0.1,0.25,0.3,0.4,0.5]).T.loc['volume','50%']
    grag_open_interest = data.describe([0.1,0.25,0.3,0.4,0.5]).T.loc['open_interest','40%']
    grag_open_interest2 = data.describe([0.1,0.25,0.3,0.4,0.5]).T.loc['open_interest','50%']
    print(e,grag_volume,grag_open_interest,grag_open_interest2)
    data1 = data.copy()
    for i in data1.index:
        if data1.loc[i,'open_interest']>30000:
            start1 = i
            break
    print(start1)
    for i in data1.index[::-1]:
        if 'MA2201' in e:
            end1 = len(data1)
            break
        if data1.loc[i,'open_interest']<60000 and data1.loc[i-1,'open_interest']>60000:
            end1 = i
            break
    print(end1)
    data2 = data1.iloc[start1:end1,:]
    print(data2,len(data))
```

* 经测试，全部改为持仓量数据更优
* 起始为open_interest>30000,终止为open_interest<60000
