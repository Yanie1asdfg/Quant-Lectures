对于回归问题和分类问题都有哪些评估指标

**分类学习任务中的一些评估指标**
* 二分类问题
    * 混淆矩阵confusion matrix
    * 准确度 accuracy
    * 精确率 precision
    * 召回率 recall
    * F1值
    * ROC曲线
    * AUC 面积
* 多分类
    * 宏平均 先在各个混淆矩阵中分别计算出结果，再计算平均值
    * 微平均 将二元混淆矩阵的对应的元素进行平均，得到 TP、TN、FP、FN 的平均值，然后再根据这些平均值来计算，这种方式称为“微平均”。

**回归学习任务中的一些评估指标**  

| 误差 | 表示
| ------ | ------ |
| 均方误差 | MSE |
| $R^2$ | r2 |
| 平均绝对误差 | MAE |
| 均方根误差 | RMSE |

# 均方误差mse

MSE的本质，其实是样本真实数据与回归结果的差异

获取回归树的结果时，我们往往选择均方误差作为我们的评估（在分类树中这个指标是score代表的预测准确率）

在回归中，我们追求的是，MSE越小越好。
