$$Loss(w,b)=\sum_{n=1}^{n}\left ( \hat{y}^{n}-(b+w\cdot x_{cp}^{n})\right )^{2}$$

pick the best function

$$f^{*}=\arg \min \limits_{f}L(f)$$

$$w^{*},b^{*}=\arg \min \limits_{w,b}L(w,b)=\arg \min \limits_{w,b}\sum_{n=1}^{n}\left ( \hat{y}^{n}-(b+w\cdot x_{cp}^{n})\right )^{2}$$

$\arg\min$的意思使后面的函数取到最小值

## **Gradient Descent**

$L(w)$需要可微分

$f^{*}=\arg \min \limits_{w}L(w)$

推广到两个参数
