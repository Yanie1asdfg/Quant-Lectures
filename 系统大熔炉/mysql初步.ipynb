Python操作MySQL的步骤如下：

> 建立数据库连接  
> 通过 cursor() 创建游标对象（pymysql通过游标来执行sql和获取结果）  
> 使用 execute() 执行sql    
> 获取结果（查询）/提交事务（增删改）    
> 关闭游标    
> 关闭连接  

* net start mysql 启动 mysql 服务    
* net stop mysql 关闭 mysql 服务   
* mysql -hlocalhost -uroot -proot 进入mysql数据库，其中-h表示服务器名，localhost表示本地；  -u为数据库用户名，root是mysql默认用户名；-p为密码，如果设置了密码，可直接在-p后直接输入。   
* 输入show databases；显示你有的数据库（mysql数据库中的命令必须以分号结尾“；”）；
* use databases名字；选择要操作的数据库；
* show tables; 显示所选数据库中的表；
* 如果要退出mysql数据库，输入exit;回车。
* 查看数据库数据存放的位置  show global variables like "%datadir%";

* CREATE DATABASE test_db;
* CREATE DATABASE IF NOT EXISTS test_db;
* drop database shfe;
* DROP TABLE table_name ;   
* SELECT concat('DROP TABLE IF EXISTS ', table_name, ';')  一次删除所有的表

```python
##将数据写入mysql的数据库，但需要先通过sqlalchemy.create_engine建立连接,且字符编码设置为utf8，否则有些latin字符不能处理  

# 第一个参数thedataframe是需要导入的pd dataframe,

# 第二个参数tablename是将导入的数据库中的表名

# 第三个参数yconnect是启动数据库的接口，pd 1.9以后的版本，除了sqllite，均需要通过sqlalchemy来设置

# 第四个参数databasename是将导入的数据库名字

# 第五个参数if_exists='append'的意思是，如果表tablename存在，则将数据添加到这个表的后面

yconnect = create_engine('mysql+pymysql://root:0000@localhost:3306/futures_c?charset=utf8mb4')  #
pd.io.sql.to_sql(df,'shfe_cu', yconnect, schema='futures_c' , index = False, index_label=None,if_exists='replace')
# df.to_sql('shfe_cu', yconnect,index = False, if_exists='replace')# index_label=None,
```

```python
import pymysql

#打开数据库连接
conn = pymysql.connect(
        host='localhost',   # 本地服务器
        user='root',
        password='0000',  # 你的数据库密码
        port=3306,          # 默认端口
        charset = 'utf8',
        db = 'futures_c')
#创建游标对象
cursor = conn.cursor()

#sql预处理语句之创建表格
sql ="""CREATE TABLE IF NOT EXISTS shfe_l
    (code VARCHAR(30) NOT NULL,
     date1 VARCHAR(30) NOT NULL ,
     date2 VARCHAR (30) NOT NULL ,
     date3 VARCHAR (30) NOT NULL ,
     date4 VARCHAR (30) NOT NULL ,
     price INT(30) ,
     name VARCHAR (30) NOT NULL ,
     exchange VARCHAR (30) NOT NULL)"""

cursor.execute(sql)

#sql预处理语句之往表格中插入数据
sql = """ALTER TABLE shfe_l CONVERT TO CHARACTER SET utf8mb4"""

cursor.execute(sql)

#提交到数据库执行，代表一个事务的结束
conn.commit()

#关闭数据库
conn.close()
```
