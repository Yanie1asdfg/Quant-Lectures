#### **美国宽松的货币政策具体是指？**

1.指大量印发钞票，增加政府的财政支出，扩大市场上货币的流通量，促进经济发展。     
2.这种政策会使美元贬值，美国国债的收益减少，对大量持有美国国债的投资者｛中国是美国国债的海外最大持有者｝造成巨大的损失，美国继续实行美元泛滥的举措，为的就是稀释各国对美债权。以便在汇率反差上吃掉他国拥有的美国债。(存疑)  
3.热钱流入股市和房市而非债市，国债流动性减少，收益降低(存疑)  
4.受货币政策上的预期变动而推动。其表现是热钱大量涌进A股，促使煤炭、有色等具保值性投资的资源类股票大幅暴涨，当然这也是因为全球通胀预期及国际商品价格大幅上扬所造成的。   
5.美元的大幅贬值，这将意味着黄金、原油以及大宗商品的价格暴涨。

#### **量化宽松(QE：Quantitative Easing)**    

1.是指中央银行在零利率或近似零利率时，通过购买国债等中长期债券，增加基础货币供给，向市场注入大量流动性资金的干预方式，以鼓励开支和借贷。量化指扩大一定数量的货币发行，宽松即减少银行的资金压力。   
2.中央银行利用凭空创造出来的钱在公开市场购买政府债券、借钱给接受存款机构、从银行购买资产等。这些都有助降低**新增政府债券的收益率**和降低银行同业隔夜利率，**可以避免存量国债市场价值显著缩水**   
3.它是一把双刃剑，一方面有助于缓解市场资金紧张状况，经济恢复增长；但长期来看，埋下通胀的隐患，在经济增长停滞的情况下，或许会引起滞涨。

#### **广义货币供应量（M2）**

指流通于银行体系之外的现金加上企业存款、居民储蓄存款以及其他存款，它包括了一切可能成为现实购买力的货币形式，通常反映的是社会总需求变化和未来通胀的压力状态。   
货币（M0）=流通中的现金，即流通于银行体系之外的现金。   
狭义货币（M1）=（M0）+活期存款；   
广义货币（M2）=M1+准货币（单位定期存款+居民定期存款+其他存款+证券公司客户保证金+住房公积金中心存款+非存款类金融机构在存款类金融机构的存款）；    
M3=M2+其他短期流动资产（如国库券、银行承兑汇票、商业票据等）    
经济学所说的货币通常是指M2，M2包括准货币。

#### **定义通胀**

CPI 和 PPI 常用来衡量物价指数，前者从消费者角度，后者从生产者角度，若用CPI同比增长率和PPI同比增长率表示物价变化，通胀时段可以定义为：
1.CPI同比增速突破2%，最高超过5%，从超过2%至顶点的时间段   
2.或者，PPI同比增速突破2%，最高超过5%，从超过2%至顶点的时间段 

注：   
同比就是比去年同期，环比就是比上个月    
同比发展速度主要是为了消除季节变动的影响，用以说明本期发展水平与去年同期发展水平对比而达到的相对发展速度。环比发展速度是报告期水平与前一时期水平之比，表明先行时期的发展速度。

#### **通胀时期的应对**

1.商品在通胀时期的表现最优，平均能获得14%的年化收益，胜率100%，从各个板块来看，能源和工业品表现最为亮眼，并且胜率很高；黄金等贵金属表现也不错，胜率略低。  
2.传统的债券和股票在通胀时期并不待见，平均收益率为负，胜率很低，对常见的资产配置带来挑战；房地产表现一般，没有想象的如日中天，8次通胀里只有2次收益为正   
3.股票因子模型表现差异较大，动量表现尚可，胜率也很高，规模和价值就平淡无奇；   
4.趋势跟踪表现非常惊艳，得益于通胀时期资产价格的高波动，能获得25%的年化，胜率100%，CPI视角下只有长周期模型胜率较高，PPI视角下效果会好点。  
4.资产配置模型表现优异，表明分散化和多样化是抵挡通胀风险的良药
