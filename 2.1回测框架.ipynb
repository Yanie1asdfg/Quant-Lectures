### 6.回测指标分析

> 收益指标  
> 风险指标

#### 6.1 收益指标

策略回测收益率

基准收益率

> 将全部的资金在相同的时间买入参照标的，回测得出的收益情况, 最常用的是沪深300

策略年化回测收益率
> 总的收益比较没有多大意义，一般比较年化收益

最大回测率

> 最大回撤率是指在选定周期内任一历史时点往后推，产品净值走到最低点时的收益率回撤幅度的最大值。 这个指标特别重要   
> 描述可能出现的极端亏损情况   
> 最大回测越小越好

计算：  
D为某一天的净值，i为某一天，j为i后的某一天，Di为第i天的产品净值，Dj则是Di后面某一天的净值    
drawdown=max（Di-Dj）/Di，其实就是对每一个净值进行回撤率求值，然后找出最大的。  

夏普比率

  
> 策略收益波动率：投资组合的标准差  
> 每承受一单位总风险，会产生多少的超额报酬  
> 夏普比率越大，说明单位风险所获得的的风险回报越高
