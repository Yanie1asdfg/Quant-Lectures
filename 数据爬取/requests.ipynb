```python
import requests
url = 'http://www.baidu.com'
res = requests.get(url)
print(res.encoding)
print(res.headers)  # 返回如果没有Content-Type，encoding='utf-8',如果有charset,就以设置为准，其他以res.encoding返回为准
print(res.url)
```

```python
print(res.text)
res.encoding = 'utf-8'
print(res.text)
```

```python
import requests
url = 'http://www.dianping.com'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36 Edg/89.0.774.63"}

res = requests.get(url,headers = headers)
print(res.encoding)
print(res.headers)  # 返回如果没有Content-Type，encoding='utf-8',如果有charset,就以设置为准，其他以res.encoding返回为准
print(res.url)
print(res.status_code)
```

```python
print(res.text)
```
